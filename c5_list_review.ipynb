5. 已知
```
fruits = ['apple', 'banana', 'orange', 'banana', 'pear']
```
如何只找到 第二个 'banana' 的 index？

````python
fruits = ['apple', 'banana', 'orange', 'banana', 'pear']
first = fruits.index('banana')
fruits.index('banana', first + 1)
````

````python
fruits = ['apple', 'banana', 'orange', 'banana', 'pear']
fruits.index('banana', fruits.index('banana') + 1)
# 这个写法的问题是，你需要保证第一个index必须成功返回
````

````python
# 所以最好的版本是：
fruits = ['apple', 'banana', 'orange', 'banana', 'pear']

try:
    first = fruits.index('banana')
    second = fruits.index('banana', first + 1)
except ValueError:
    second = None
print(second)
````

8. 上题如何真正“互不影响”？

````python
a = [[1, 2], [3, 4]]

import copy
b = copy.deepcopy(a)
b[0].append(99)

print(a)
print(b)
````

````python
# 另外一个方法（只对二维list有效）

a = [[1, 2], [3, 4]]

b = [x.copy() for x in a]
b[0].append(99)

print(a)
print(b)
````

10. 你从设备拉回一个 list，里面可能有重复 hostname：<br>
```
hosts = ['r1', 'r2', 'r1', 'r3', 'r2']
```
要求：
- 保持原顺序
- 去重
- 结果还是 list

````python
hosts = ['r1', 'r2', 'r1', 'r3', 'r2']
host_unique = []
for i in hosts:
    if i not in host_unique:
        host_unique.append(i)
print(host_unique)
````

````python
# 还可以用dict.fromkeys()

hosts = ['r1', 'r2', 'r1', 'r3', 'r2']

hosts_unique = list(dict.fromkeys(hosts))

print(host_unique)
````

11. 不用 for-loop，把下面操作写成一行：
- 清空 list
- 反转 list
- 把 list 里所有元素都删掉但保留对象本身（in-place）

````python
a = [1, 2, 3]
print(id(a))
a.clear()
print(a)
print(id(a))
````

````python
a = [1, 2, 3]
print(id(a))
a.reverse()
print(a)
print(id(a))
````

````python
a = [1, 2, 3]
print(id(a))
a = a[::-1]
print(a)
print(id(a))
````

````python
a = [1, 2, 3]
print(id(a))
a[:] = []
print(a)
print(id(a))
````

````python
a = [1, 2, 3]
print(id(a))
del a[:]
print(a)
print(id(a))
````

````python
a = [1, 2, 3]
print(id(a))
a = []
print(a)
print(id(a))
````
