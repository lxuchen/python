1. 给定: 
```
nums = [1, 2, 3, 4, 5, 6]
```
用 一行 list comprehension 生成一个新 list，要求：
- 只保留偶数
- 每个元素平方

````python
nums = [1, 2, 3, 4, 5, 6]
result = [x**2 for x in nums if x%2 == 0]
print(result)
````

2. 给定：
```
nums = [-3, -1, 0, 2, 4]
```
生成一个新 list：
- 正数 → 保留原值
- 非正数 → 变成 0

````python
nums = [-3, -1, 0, 2, 4]
result = [x if x > 0 else 0 for x in nums ]
print(result)
````

4. 给定：
```
devices = ['r1', 'r2', 'r3']
ips = ['10.0.0.1', '10.0.0.2', '10.0.0.3']
```
zip成：
```
['r1:10.0.0.1', 'r2:10.0.0.2', 'r3:10.0.0.3']
```

````python
devices = ['r1', 'r2', 'r3']
ips = ['10.0.0.1', '10.0.0.2', '10.0.0.3']
result = [f"{d}:{ip}" for d, ip in zip(devices, ips)]
print(result)
````

5. 给定：
```
a = [1, 2, 3, 4]
b = [10, 20]
```
问题：
- list(zip(a, b)) 的结果是什么？
- 用一行 list comprehension 计算：
```
a[i] + b[i]
```
（只算能配对的部分）

````python
a = [1, 2, 3, 4]
b = [10, 20]
print(list(zip(a, b)))
result = [x + y for x, y in zip(a, b)]
print(result)
````

6. 给定：
```
pairs = [(1, 10), (2, 20), (3, 30)]
```
不用 for-loop，生成两个 list：
```
xs = [1, 2, 3]
ys = [10, 20, 30]
```

````python
pairs = [(1, 10), (2, 20), (3, 30)]
unpack = list(zip(*pairs))
print(unpack)
xs = list(unpack[0])
ys = list(unpack[1])
print(xs)
print(ys)

# 更好的版本

xs, ys = zip(*pairs)
print(list(xs))
print(list(ys))
````

7. 给定：
```
devices = [
    {'name': 'r1', 'latency': 20},
    {'name': 'r2', 'latency': 5},
    {'name': 'r3', 'latency': 15}
]
```
用一行代码：
- 按 latency 从小到大排序
- 返回排序后的 list of dict


````python
devices = [
    {'name': 'r1', 'latency': 20},
    {'name': 'r2', 'latency': 5},
    {'name': 'r3', 'latency': 15}
]
result = sorted(devices, key=lambda x: x['latency'])
print(result)
````

9. lambda + map（故意刁钻）
给定：
```
nums = [1, 2, 3, 4]
```
用 map + lambda 得到：
```
[1, 4, 9, 16]
```
然后回答（不用写代码）：
- 这个题用 list comprehension 会不会更好？为什么？

````python
# map + lambda

nums = [1, 2, 3, 4]
result = list(map(lambda x: x**2, nums))
print(result)
````

````python
# List comprehension

nums = [1, 2, 3, 4]
result = [x**2 for x in nums]
print(result)

# of cause list comprehension is better
````

10. 综合题（真实 automation 风格）
给定：
```
hosts = ['r1', 'r2', 'r3']
cpu = [30, 85, 55]
```
生成一个新 list，只包含：
- CPU > 50 的主机
- 格式："rX:YY%"

结果示例：
```
['r2:85%', 'r3:55%']
```
要求：
- 用 zip
- 用 list comprehension
- 不用显式 for-loop

````python
hosts = ['r1', 'r2', 'r3']
cpu = [30, 85, 55]
result = [f"{r}:{c}%" for r, c in zip(hosts, cpu) if c > 50]
print(result)
````
