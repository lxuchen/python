## 2.1 Invoking the Interpreter

- `python` # depends on the environment (venv is important!)
- Ctrl-D (Ctrl-Z on Windows) or `exit()` or `quit()`
- command line editing (Ctrl-P to verify). To be honest, it does not matter too much.
- another way to start the interpreter: `python -c command [arg] ...`
- use python modules as scripts: `python -m module [arg] ...`
- drop into interactive mode after script: `python -i`

### 2.1.1 Argument Passing
- the script name and additional arguments thereafter are turned into a list of strings and assigned to the `argv` in the `sys` module

### 2.2.1 Source Code Encoding

- by default utf-8
- to declare an encoding other than the default, a special comment line should be added as the first line of the file:

```
# -*- coding: gb2312 -*-
```

Except if the file starts with a Unix `shebang` line, in this case, the encoding declaration should be added as the 2nd line of the file:

```
#!/usr/bin/env python3
# -*- coding: gb2312 -*-
```
