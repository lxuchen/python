9. 网络工程风格题
```
sw1_ports = ["eth0", "eth1", "eth2", "eth3"]
sw2_ports = ["eth2", "eth3", "eth4"]
```
找出：
- 两台设备共有的接口
- 只在 sw1 上存在的接口




````python
sw1_ports = ["eth0", "eth1", "eth2", "eth3"]
sw2_ports = ["eth2", "eth3", "eth4"]
print(set(sw1_ports) & set(sw2_ports))
print(set(sw1_ports) - set(sw2_ports))
````
