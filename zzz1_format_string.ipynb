快速过一遍format string的内容：

⸻

1️⃣ f-string 是首选（Python 3.6+）

基本形式：
```
name = "r1"
ip = "10.0.0.1"
s = f"{name}:{ip}"
```
规则：
- 前缀 f
- {} 里放 任意表达式

⸻

2️⃣ 表达式，不只是变量（很重要）
```
f"{x+1}"
f"{x*2}"
f"{x.upper()}"
f"{a if a > 0 else 0}"
```
你在 list comprehension / zip 里用的基本都是这类。

⸻

3️⃣ 常用格式控制（必须记住的 5 个）

数字
```
f"{n:d}"        # 整数
f"{n:.2f}"      # 浮点保留 2 位
f"{n:06d}"      # 补零
```
例：
```
n = 7
f"{n:03d}"   -> "007"
```

⸻

百分比
```
x = 0.8532
f"{x:.1%}"   -> "85.3%"
```
（这在 metric / automation 输出里非常常见）

⸻

对齐（日志用）
```
f"{name:<10}"   # 左对齐
f"{name:>10}"   # 右对齐
f"{name:^10}"   # 居中
```

⸻

4️⃣ dict / list 解包（你会经常用）
```
d = {"name": "r1", "ip": "10.0.0.1"}
f"{d['name']}:{d['ip']}"
```

⸻

5️⃣ f-string + zip / comprehension（你刚刚用到的）
```
[f"{d}:{ip}" for d, ip in zip(devices, ips)]
```
这是 黄金组合。

⸻

6️⃣ 调试神器（Python 3.8+）
```
f"{x=}"
```
等价于：
```
"x=10"
```
例：
```
x = 10
f"{x=}"   -> "x=10"
```
日志里非常好用。

⸻

7️⃣ 不推荐但你要认识的

老写法（不建议新代码用）：
```
"{}:{}".format(name, ip)
"%s:%s" % (name, ip)
```
记住就行，不要主动写。

⸻

最终工程结论（记住这 4 条）
- 新代码：只用 f-string
- {} 里可以放 表达式
- 日志 / automation 输出 = f-string + 对齐
- 和 list comprehension / zip 是天然搭配


1. 给定：
```
host = "r1"
ip = "10.0.0.1"
```
用 f-string 生成：
```
"r1(10.0.0.1)"
```

````python
host = "r1"
ip = "10.0.0.1"
result = f"{host}({ip})"
print(result)
````

2. 表达式 + 条件

给定：
```
cpu = 87
```
用 一行 f-string 生成：
- CPU ≥ 80 → "HIGH:87%"
- 否则 → "OK:87%"

````python
cpu = 87
result = f"{'HIGH' if cpu >= 80 else 'OK'}:{cpu}"
print(result) 
````

3. 数字格式（工程常用）

给定：
```
latency = 3.45678
```
生成字符串：
```
"latency=3.46ms"
```
要求：
- 四舍五入
- 保留两位小数



````python
latency = 3.45678
result = f"latency={latency:.2f}ms"
print(result)
````

4. 补零（ID / 设备号）

给定：
```
i = 7
```
生成：
```
"eth007"
```

````python
i = 7
result = f"eth{i:03d}"
print(result)
````

5. 对齐（日志输出）

给定：
```
name = "r1"
status = "UP"
```
生成一个字符串，使得：
- name 左对齐，占 6 位
- status 右对齐，占 4 位

肉眼效果类似：
```
"r1     |   UP"
```

````python
name = "r1"
status = "UP"
result = f"{name:<6}|{status:>4}"
print(result)
````

给定：
```
device = {"name": "r2", "cpu": 92}
```
生成：
```
"r2 cpu=92%"
```

````python
device = {"name": "r2", "cpu": 92}
result = f"{device['name']} cpu={device['cpu']}%"
print(result)
````

7. f-string + zip + comprehension（核心）

给定：
```
hosts = ["r1", "r2", "r3"]
cpu = [30, 85, 55]
```
生成一个 list，只包含 CPU ≥ 50 的项，格式：
```
["r2:85%", "r3:55%"]
```
要求：
- 必须用 f-string
- 必须用 zip
- 一行 list comprehension

````python
hosts = ["r1", "r2", "r3"]
cpu = [30, 85, 55]


# results = [x for x in zip(hosts, cpu) if x[1] >= 50]
# list_result = []
# for i in results:
#     list_result.append(f"{i[0]}:{i[1]}%")
# print(list_result)

result = [f"{h}:{c}%" for h, c in zip(hosts, cpu) if c >=50]
print(result)
````

8. f-string 调试格式（Python 3.8+）

给定：
```
x = 10
y = 20
```
生成字符串（不手写变量名）：
```
"x=10, y=20"
```

````python
x = 10
y = 20
result = f"{x=}, {y=}"
print(result)
````

9. lambda + sorted + f-string（面试级）

给定：
```
devices = [
    {"name": "r1", "latency": 20},
    {"name": "r2", "latency": 5},
    {"name": "r3", "latency": 15},
]
```
要求：
- 按 latency 从小到大排序
- 生成 list：
```
["r2(5ms)", "r3(15ms)", "r1(20ms)"]
```

````python
devices = [
    {"name": "r1", "latency": 20},
    {"name": "r2", "latency": 5},
    {"name": "r3", "latency": 15},
]
result = [f"{d['name']}({d['latency']}ms)" for d in sorted(devices, key=lambda x: x['latency'])]
print(result)
````

10. 综合（非常真实的 automation 输出）

给定：
```
hosts = ["r1", "r2", "r3"]
ips = ["10.0.0.1", "10.0.0.2", "10.0.0.3"]
status = [True, False, True]
```
生成：
```
["r1 10.0.0.1 UP", "r3 10.0.0.3 UP"]
```
规则：
- status 为 False 的丢弃
- True → "UP"
- False → "DOWN"
- 必须用 f-string + zip + list comprehension


````python
hosts = ["r1", "r2", "r3"]
ips = ["10.0.0.1", "10.0.0.2", "10.0.0.3"]
status = [True, False, True]

# result = [f"{h} {i} {'UP' if s else 'DOWN'}" for h, i, s in zip(hosts, ips, status) if s]

result = [
    f"{h} {ip} UP"
    for h, ip, s in zip(hosts, ips, status)
    if s
]

print(result)
````
