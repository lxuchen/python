```python
data = [
    {"device": "r1", "status": "up"},
    {"device": "r2", "status": "down"},
    {"device": "r1", "status": "down"},
    {"device": "r3", "status": "down"},
    {"device": "r2", "status": "up"},
]

# 任务：
# 统计每个 device 的 up / down 次数
# 目标输出：
# {
#     "r1": {"up": 1, "down": 1},
#     "r2": {"up": 1, "down": 1},
#     "r3": {"up": 0, "down": 1}
# }

# 要求：
# 1) 不能写两次循环
# 2) 所有 device 必须出现（即使某个状态为 0）
# 3) 不能 hardcode device 名字
# 4) 不用 Counter
# 5) 用 dict 实现
#
# 提示：嵌套 dict 初始化（比如 setdefault 或 get）

updown_counts = {}
for entry in data:
    device = entry["device"]
    status = entry["status"]
    if device not in updown_counts:
        updown_counts[device] = {"up": 0, "down":0}
    updown_counts[device][status] = updown_counts[device][status] +1
    # updown_counts[device][status] = updown_counts[device].get(status, 0) + 1
    # updown_counts[device][status] = updown_counts.get(device, {}).get(status, 0) + 1
print(updown_counts)
```
