8. 给定多个设备：
```
devices = (
    ("r1", "10.0.0.1"),
    ("r2", "10.0.0.2"),
    ("r3", "10.0.0.3"),
)
```
用 for 循环 unpack 并打印：
```
r1 -> 10.0.0.1
```

````python
devices = (
    ("r1", "10.0.0.1"),
    ("r2", "10.0.0.2"),
    ("r3", "10.0.0.3"),
)
for device in devices:
    r, i = device
    print(f"{r} -> {i}")

# Better version:

for r, i in devices:
    print(f"{r} -> {i}")
````

9. 为什么 tuple 可以作为 dict 的 key，而 list 不行？

# tuple is immutable

10.	什么情况下你会用 tuple 而不是 list？
从“工程实践角度”回答，不要只说“immutable”。

# 不知道诶！

好，这题我直接讲底层逻辑。

```
在 Python 里，dict 的 key 必须满足两个条件：
	1.	必须是 hashable
	2.	hash 值在生命周期内不能改变

dict 底层是哈希表。
查找流程是：

hash(key) → 定位 bucket → 再用 == 比较确认

如果一个对象的内容可以改变，而 hash 又依赖于内容，就会出大问题。

举个逻辑例子（假设 list 允许做 key）：

l = [1, 2]
d = {l: "router"}

dict 会计算：

hash(l)

如果你之后：

l.append(3)

那么 l 的内容变了。

如果 hash 是基于内容计算的，那么：

hash(l) 也应该变

但 dict 里它已经存放在原来的 bucket 里了。

这会导致：
	•	你再用 l 查找时找不到
	•	哈希表结构被破坏
	•	数据一致性崩溃

这就是为什么：

可变对象不能是 key

⸻

为什么 tuple 可以？

因为：
	•	tuple 是 immutable
	•	tuple 的 hash 在创建时可确定
	•	tuple 内部如果全部元素也是 hashable，那么 tuple 就是 hashable

注意：

(1, 2, 3)      → hashable
(1, [2, 3])    → 不 hashable

第二个不行，因为里面有 list。

Python 在计算 tuple 的 hash 时，会递归检查元素是否 hashable。

⸻

工程思维总结：

dict key 的核心不是 immutable
而是：

“hash 值必须稳定”

immutable 只是实现稳定 hash 的一种方式。

```
