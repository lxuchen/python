Generator 就是一种 可以逐步产出数据的迭代器。

普通函数：
- 一次执行完
- 返回一个结果

Generator 函数：
- 每次只产出一个值
- 可以暂停
- 下次从暂停处继续

```python
def count():
    yield 1
    yield 2
    yield 3

g = count()
print(g)
```

```python
for x in count():
    print(x)
```

```python
g = count()
print(next(g))
print(next(g))
print(next(g))
```

Generator vs list:
- list 一次性算完，存在内存中
- generator 按需生成，不占用大内存

```python
num = [x*x for x in range(5)]
print(num)
```

```python
num = (x*x for x in range(5))
print(next(num))
print(next(num))
print(next(num))
print(next(num))
print(next(num))

```

那么，什么是yield？

yield就是会把一个函数变成 generator（生成器）函数。

```python
# 普通函数，返回值就拉倒

def f():
    return 1

print(f())
```

```python
# 带yield的函数，返回一个Generator对象，必需通过for loop调用
def y():
    yield 1
    yield 2
    yield 3

print(y())
for x in y():
    print(x)
```

```python
#或者：

x = y()
print(next(x))
```

```python
# 还能记住状态
print(next(x))
```

那么到底yield做了什么？

普通 return：
- 执行
- 返回
- 函数结束

yield：
- 暂停函数
- 把当前值“产出”
- 保留函数状态
- 下次从暂停位置继续执行

它本质是：

可暂停函数执行的机制。

为什么需要yield？

- 节省内存
