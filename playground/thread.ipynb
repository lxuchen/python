Thread = 同一进程里并行执行的多条执行路径。

简单理解就是：

一个 Python 程序里可以同时做多件事。

```python
# 基础code

import threading
import time

def task():
    print("start")
    time.sleep(2)
    print("end")

t = threading.Thread(target=task)
t.start()
```

```python
# 典型多线程：

import threading
import time

def task(n):
    print("task", n, "start")
    time.sleep(2)
    print("task", n, "end")

threads = []

for i in range(3):
    t = threading.Thread(target=task, args=(i,))
    t.start()
    threads.append(t)

for t in threads:
    t.join()
```

现在讲最关键的一点（Python 特有）：

GIL（Global Interpreter Lock）

这个是 Python 线程绕不开的东西。

简单理解：

同一时刻只能有一个线程执行 Python bytecode。

所以：

线程 ≠ CPU 并行计算。

那为什么还用 thread？

因为很多任务不是 CPU-bound，而是 IO-bound：

比如：
	•	网络请求
	•	文件读写
	•	数据库访问
	•	API polling
	•	SSH/network automation

当线程在等 IO 时：

GIL 会释放，其他线程能跑。

所以对 automation / 网络测试特别有用

现在对比 async 和 thread（这个很关键）。

async：
	•	单线程
	•	event loop 调度
	•	需要 async/await 支持
	•	更轻量

thread：
	•	多线程
	•	OS 调度
	•	可以直接跑同步代码
	•	简单但稍重

你现在工作场景（IBNT、network automation）：

很多 legacy SDK、CLI tools、设备接口都是同步阻塞的。

这时 thread 比 async 更现实。
