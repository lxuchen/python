实战常用 的 string methods

---

一、清洗 & 预处理（最常用）

1️⃣ strip() / lstrip() / rstrip()

去掉 CLI 输出里的空格和换行。
```
line = "  Gig0/0 up  \n"
line = line.strip()
```

典型场景：
	•	解析 show interface
	•	处理设备回显
	•	读取文件行



2️⃣ lower() / upper()

做大小写标准化。

status = status.lower()
if status == "up":

网络设备经常有：
	•	up / UP / Up
	•	connected / CONNECTED

必须 normalize。



3️⃣ replace()

清理脏字符。

output = output.replace("\r", "")

常见：
	•	去掉 \r
	•	去掉 ,
	•	去掉特殊符号

---

二、查找 & 判断

4️⃣ startswith() / endswith()

比 if "Gig" 更安全。

if line.startswith("Gig"):

用于：
	•	判断 interface 类型
	•	过滤 header 行


5️⃣ in

if "down" in line:

最常用模糊匹配。



6️⃣ find()

index = line.find("up")

很少直接用，一般 in 就够了。

---

三、拆分 & 解析（核心）

7️⃣ split() ⭐⭐⭐

网络自动化最重要。

parts = line.split()

默认按空格切。

例：

Gig0/0    up    10.0.0.1

iface, status, ip = line.split()

也可以：

line.split(",")

用于 CSV / structured 输出。



8️⃣ splitlines()

解析 multi-line CLI 输出。

lines = output.splitlines()

比 output.split("\n") 更干净。

⸻

四、拼接 & 构造输出

9️⃣ join() ⭐⭐⭐

日志输出常用。

",".join(interface_list)

network automation 典型场景：

print(" ".join(commands))



1️⃣0️⃣ f-string（不是 method，但必须会）

cmd = f"interface {iface}"

写 config automation 时必须熟练。

---

五、验证类（很实用）

1️⃣1️⃣ isdigit()

if vlan_id.isdigit():



1️⃣2️⃣ isalpha() / isalnum()

校验 device name。

---

六、格式化

1️⃣3️⃣ format()

现在多用 f-string，但 legacy code 常见：

"interface {}".format(iface)


---

七、真实 Network Automation 例子

解析 CLI：

output = """
Gig0/0 up 10.0.0.1
Gig0/1 down 10.0.0.2
"""

for line in output.splitlines():
    line = line.strip()
    if not line:
        continue

    iface, status, ip = line.split()

    if status.lower() == "down":
        print(f"{iface} is down")

output parsing 本质都是这些操作。

⸻

你在网络自动化里真正会频繁用的就这几个：
	•	strip
	•	split
	•	splitlines
	•	lower
	•	replace
	•	startswith
	•	join
	•	f-string

其他的属于锦上添花。

````python
output = """Gig0/0 up 10.0.0.1 
Gig0/1 down 10.0.0.2 
Gig0/2 UP 10.0.0.3
"""


# 找出所有 down 接口（大小写不敏感）
# 输出接口名列表

lines = output.splitlines()
down_interfaces = []
for line in lines:
    interface_name, interface_status, interface_ip = line.split()
    if interface_status.lower() == 'down':
        down_interfaces.append(interface_name)
print(down_interfaces)
````

````python
output = "  VLAN 100 active \r\n"

# 去掉所有首尾空白和 \r
# 最终得到 "VLAN 100 active"

print(output.strip())
````

````python
output = "hostname   core-sw-01"

# 提取 hostname 字符串
# 不允许用 regex，只用 string method

print(output.split()[1])
````

````python
output = "Gig0/1,Gig0/2,Gig0/3,Gig0/4"

# 转成 list
# 再重新拼成 "Gig0/1 Gig0/2 Gig0/3 Gig0/4"

print(" ".join(output.split(",")))
````

````python
output = """interface Gig0/1
 switchport mode trunk
"""

# 判断该接口是不是 trunk
# 不用 regex

lines = output.splitlines()
print(lines)
if 'trunk' in lines[1]:
    print(f"Yes, {lines[0]} is a trunk interface!")
else:
    print(f"No, {lines[0]} is not a trunk interface!")
````

````python
output = """Neighbor 10.1.1.1 Established
Neighbor 10.1.1.2 Idle
Neighbor 10.1.1.3 established"""

# 找出所有 Established neighbor IP
# 忽略大小写

log = output.splitlines()
established_neighbors = []
for log_entry in log:
    _, ip_address, bgp_state = log_entry.split()
    if bgp_state.lower() == "established":
        established_neighbors.append(ip_address)
print(established_neighbors)
````
