async = 让程序在等待 IO 时可以干别的事。

不是更快 CPU 计算，而是减少“干等”。

````python
import time

def task():
    print("start")
    time.sleep(2)
    print("end")

task()

# 这里：CPU 完全闲着 2 秒什么也没干, 如果同时有 100 个任务就会卡死。
````

async 的思路是：

“等的时候先去做别的任务”。

比如：
- 等网络返回
- 等磁盘 IO
- 等设备响应
- 等 API response

这些都叫：

IO-bound task。

````python
# Python async 写法长这样：
import asyncio

async def task():
    print("start")
    await asyncio.sleep(2)
    print("end")

task()
````

几个关键点：

① async def

表示：

这是一个 coroutine（协程函数）。

调用不会立刻执行：

```
task()
```
返回 coroutine 对象。

类似 generator 的感觉。

② await

意思是：

“这个地方我要等一下，但允许别的任务先跑。”

非常像 yield 的升级版。

③ 运行 async 代码

需要 event loop：
```
asyncio.run(task())
```
这是入口。

````python
# 现在看 async 真正的威力。

# 同步，需要6秒

import time

def task():
    print("start")
    time.sleep(2)
    print("end")

task()
task()
task()

````

````python
# 异步，只要2秒

import asyncio

async def task():
    print("start")
    await asyncio.sleep(2)
    print("end")

async def main():
    result = await asyncio.gather(task(), task(), task())


await main()
````
