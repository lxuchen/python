## 学习一下map, filter 以及generator expression

### `map` - 批量变换，把一个函数应用到每个元素上：

```
map(function, iterable)
```

### `filter` - 批量筛选，保留满足条件的元素：

```
filter(function, iterable)
```

### generator expression - 更节省内存的方法

```
(expression, for loop)
```

### list comprehension vs generator expression

- list comprehension: 在内存中已经完整生成了整个list
- generator expression：只有在需要的时候才生成
- 如果最后最终目标是要list，那么comprehension更好，如果不确定，用generator

比如这两行就没什么区别，comprehension反而更好：
```
print([x ** 2 for x in range(10)])
print(list(x**2 for x in range(10)))
```
但下面这个就是generator更好：
```
print(sum(x for x in range(10)))
```
因为省去了生成list的过程

### rule of thumb

- generator 只能iterate一次，iterate之后就消耗掉了
- 如果要多次使用结果，用list comprehension
- 只需要遍历一次，用generator
- 处理很多数据，用generator
- 传给join, sum, any, all这些函数，用generator，因为不需要中间生成list的步骤，只要iterable
- 如果已经有函数了，用map，如果没有现成函数，用list comprehension

````python
print(list(range(10)))
print(list(map(str, range(10))))
print(list(filter(lambda x: x % 2 == 0, range(10))))
````

````python
print([x for x in range(10)])
````

````python
print(list(x for x in range(10)))
````

````python
print(sum(x for x in range(10)))
````

````python
g = (x for x in range(10))
print(list(g))
print(list(g)) #只能iterate一次，就消耗掉了
````

````python
list(map(str, range(10)))

#有现成函数，用map很好
````

````python
[x ** 2 for x in range(10)]

#没有现成函数，用list comprehension
````

````python
list(map(lambda x: x**2, range(10)))

#否则，map里还得写lambda，可读性不高
````

````python
print(','.join(map(str, range(10))))
````
