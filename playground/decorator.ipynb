Decorator 本质就是：在不改函数源码的情况下，给函数“包一层额外行为”。

Python 用 @something 这种语法让这个过程非常自然。

```python
# 假设你有个函数：
def hello():
    print("hello")

# 现在你想每次调用这个函数前后都打印日志。最原始写法是：

def wrapper():
    print("before")
    hello()
    print("after")

wrapper()

# 但这样就改了调用方式，看起来也不优雅。
```

```python
# 用decorator：

def my_decorator(func):
    def wrapper():
        print("before")
        func()
        print("after")
    return wrapper

@my_decorator
def hello():
    print("hello")


def hi():
    print("hi")

hello()

hi = my_decorator(hi)

hi()
```
