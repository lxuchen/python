### Video #11 Slice operator

```python
# slicing assignment

a = ['apple', 'banana', 'pear']
a[1:1] = ['strawberry']
print(a)

# 这个slice如何理解？其实index本质上是元素之间的缝隙，1:1就是在apple和banana之间的那个缝隙，所以slicing总是“左闭右开”
```

# Video #13 How to read a text file

```python
file = open('test.txt', 'r')
f = file.readlines()

lst = []
for line in f:
    lst.append(line.strip())
print(lst)
file.close()
```
