什么时候用tuple?

首先，当数据“逻辑上不应该被改”时，用 tuple。
比如坐标、IP+port、RGB颜色值、配置常量这类，本质是一个固定记录，而不是会增删的数据集合。
```
server = ("10.0.0.1", 443)
color = (255, 128, 0)
```
这种如果用 list，其实是在传递一个错误信号 —— 好像它会变，但实际上不会。

---

第二种很常见：函数返回多个值时 → 默认 tuple。

Python 其实自动帮你变成 tuple：
```
def get_host():
    return "localhost", 8080
```
接收时：
```
host, port = get_host()
```
这是 Python idiomatic 写法，比 list 更标准。

---

第三种：要做 dict key 或 set 元素时。

因为 tuple immutable，而 list mutable，所以：
```
d = {}
d[(192, 168, 1, 1)] = "router"   # OK
d[[192, 168, 1, 1]] = "router"   # 报错
```
你以后做 network automation、配置 mapping 时这个会经常遇到。

---

第四种（稍微高级一点）：性能敏感、且数据固定。

tuple：
	•	占用内存略少
	•	创建略快
	•	访问略快

但这个通常不是主要理由，除非数据量非常大。

---

第五种，其实是最实用的一条经验：

👉 默认用 list，除非你明确知道数据不该改。

这是 Python 社区普遍习惯。

换句话说：
	•	不确定 → list
	•	明确固定 → tuple

---

结合你现在 Python 学习 + 工作环境（automation / network / config 类代码），我个人建议你优先关注这两个场景：
	1.	配置、标识符、坐标类 → tuple
	2.	收集数据、动态列表 → list

基本就不会错。

---

如果你愿意，我下一步可以讲一个很多人忽略但非常重要的点：

👉 “tuple 不可变 ≠ tuple 里面的对象不可变”

这个坑不少 Python 初学者都会踩。
