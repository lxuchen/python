```python
# How to append a list without using .append(x):

a = ['a', 'b', 'c', 'd']
a.append('e')
print(a)
a[len(a):] = 'f'
print(a)
```

```python
a = [1,2,3]
a.append(4)
print(a)
a.append('5')
print(a)
# a[len(a):] = 6, this is WRONG!
a[len(a):] = [6] # this is the same as .append(6)
print(a)
a[len(a):] = '7' # this is the same as .append('7')
print(a)
a[len(a):] = '12'
print(a) # note the difference
```

```python
a = [1,2,3]
a.extend([4,5])
print(a)
a.extend('67')
print(a)
```

```python
a = [1,2,3,4,5]
a.insert(0,0)
print(a)
a.insert(len(a), 6)
print(a)
```

```python
a = [1,2,3,4,5]
a.remove(3)
print(a)
a.remove(100)
```

```python
a = [1,2,3,4,5]
b = a.pop(4)
print(b)
print(a)
a.pop(100)
```

### 5.1.2 Using List as Queues

(“first-in, first-out”)



```python
a = [1,2,3,4]
a.append(5)
print(a)
print(a.pop())
print(a)
```

```python
a = [1,2,3,4]
a.append(5)
print(a)
a.pop(0)
print(a)
```

```python
from collections import deque

q = deque()
q.append(1)
q.append(2)
q.append(3)
print(q.popleft())
print(q)
```

```python
# shallow copy

a = [1, [2, 3], 4]
b = a.copy()

print(a)
print(b)
print(a is b)
print(a[1] is b[1])
b.append(5)
print(a)
print(b)
b[1].append(99)
print(a)
print(b)
```

```python
# deep copy

import copy

a = [1, [2, 3], 4]
c = copy.deepcopy(a)

print(a)
print(c)
print(a is c)
print(a[1] is c[1])
c.append(5)
print(a)
print(c)
c[1].append(99)
print(a)
print(c)
```
